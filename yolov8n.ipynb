```python
!pip install -q ultralytics opencv-python

from google.colab import files
import zipfile, os
from pathlib import Path

print("LÃ¼tfen 'train_data' klasÃ¶rÃ¼nÃ¼n ZIP halini seÃ§ (Ã¶rnek: train_data.zip):")
uploaded = files.upload()

zip_name = next(iter(uploaded))
print(f"YÃ¼klenen dosya: {zip_name}")

DATASETS_DIR = Path("/content/datasets")
DATASETS_DIR.mkdir(parents=True, exist_ok=True)

with zipfile.ZipFile(zip_name, "r") as z:
    z.extractall(DATASETS_DIR)

print("\nğŸ“ /content/datasets iÃ§eriÄŸi:")
!ls -R /content/datasets
```

```python
from pathlib import Path

DATASET_ROOT = Path("/content/datasets/train_data")

num_classes = 1
class_names = ["chicken"]  # sadece bu sÄ±nÄ±f

names_lines = "\n".join([f"  {i}: {name}" for i, name in enumerate(class_names)])

data_yaml = f"""
path: {DATASET_ROOT}
train: images/train
val: images/val
nc: {num_classes}
names:
{names_lines}
"""

with open("/content/chicken.yaml", "w") as f:
    f.write(data_yaml)

print(open("/content/chicken.yaml").read())
```

```python
from ultralytics import YOLO

# Detection modeli (kÃ¼Ã§Ã¼k ve hÄ±zlÄ±)
model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/chicken.yaml",  # az Ã¶nce oluÅŸturduÄŸun dosya
    epochs=60,                     # istersen 50-100'e Ã§Ä±karabilirsin
    imgsz=640,
    batch=4
)

results
```
